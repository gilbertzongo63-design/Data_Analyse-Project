3.Creation of at least one new analytical indicator.

```python
#Analytical indicator of the annual variance 

# Calculation of the annual variance by year"
df['variance_annuelle'] = df.groupby('Year')['average_farm_size_ha'].transform('var')

df[['Entity', 'Year', 'average_farm_size_ha', 'variance_annuelle']].head(10)
```

```python
#Analytical indicator of the relative growth of each country since the reference year (1960).

df['growth_rate_ratio'] = df['average_farm_size_ha'] / df.groupby('Entity')['average_farm_size_ha'].transform('first')

df[['Entity', 'Year', 'average_farm_size_ha', 'growth_rate_ratio']].head()
```

4.Creation of at least one meaningful visualization using Matplotlib.

```python
#Plots the kernel density estimate (KDE) of the "Year" column.
df["Year"].plot(kind="kde")
```

```python
# Counts the number of occurrences of each year in the "Year" column

df["Year"].value_counts().plot(kind="pie",legend=True)
```

```python
#Ce code permet de comparer visuellement les 10 pays ayant la plus grande taille moyenne des exploitations agricoles, en mettant en vidence les carts entre pays.
moyenne_pays = df.groupby("Entity")["average_farm_size_ha"].mean().sort_values(ascending=False)

plt.figure(figsize=(12,6))
plt.bar(moyenne_pays.head(10).index, moyenne_pays.head(10).values, color='skyblue')
plt.ylabel("Average farm size (ha)", fontsize=12)
plt.title("Ccomparison of average farm size(Top 10 pays)", fontsize=14)
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```
